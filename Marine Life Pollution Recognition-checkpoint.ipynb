### Removing doddy images

```python
import cv2 # used for computer vision
import imghdr # to check file extensions for particular images
from matplotlib import pyplot as plt # to visualize images
```

```python
data_dir = 'data' # created a variable to hold the path of data dir
```

```python
os.listdir(data_dir) # using the variable we stored path in
# it will return two folder 
```

```python
# To check every image in the folder
os.listdir(os.path.join(data_dir,'Hazard material'))
```

```python
image_exts = ['jpeg','jpg','bmp','png']  # we created standard list of extensions of images
image_exts
```

### Iterating through images and removing faulty images which doesnt match with the standard extension list we made 

```python
for image_class in os.listdir(data_dir): #looping through 'aquatic_animal' & 'hazard material' folder
    for image in os.listdir(os.path.join(data_dir,image_class)): # then we are looping through each images in the above two folders
        image_path = os.path.join(data_dir,image_class,image) # targeting individual pic from the above two iteration and saving in a variable
        try:
            img = cv2.imread(image_path) #imread allows to open up an image using opencv
            tip = imghdr.what(image_path) # this reads what iamge extension is
            if tip not in image_exts:
                print('Image not in ext list {}'.format(image_path))
                os.remove(image_path)
        except Exception as e : 
                print('Issue with image {}'.format(image_path))
```

### Split Data

```python
len(data) # 92 is the no of batches
```

```python
train_size = int(len(data)*.7)+1#data that is used for traing neural model
val_size = int(len(data)*.2) # used to evalate model while training
test_size = int(len(data)*.1) # never seen data used to test model after training
```

```python
# ALLOCATING DATA TO TRAIN , TEST , VALUATION
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```

# Building CNN Model (keras-sequential API)

```python
 #importing sequential api
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
```

```python
model = Sequential()
```

```python
#16-filters filter-size:3X3px stride=1
#input will be only ininitial layer
model.add(Conv2D(64 ,(3,3),1,activation='relu',input_shape=(256,256,3)))
model.add(MaxPooling2D()) #used to condense 

model.add(Conv2D(32,(3,3),1,activation='relu'))
model.add(MaxPooling2D())

model.add(Conv2D(32,(3,3),1,activation='relu'))
model.add(MaxPooling2D())

model.add(Conv2D(32,(3,3),1,activation='relu'))
model.add(MaxPooling2D())

model.add(Conv2D(128,(3,3),1,activation='relu'))
model.add(MaxPooling2D())

model.add(Flatten())
 
model.add(Dense(256,activation='relu')) #neural network 256-neurons
model.add(Dense(1,activation='sigmoid')) # we will get single output it will represent either 0 or 1
```

```python
model.compile('adam',loss=tf.losses.BinaryCrossentropy(),metrics=['accuracy'])
# adam-optiizer(there are different too)
```

```python
model.summary()
```

## Train

```python
logdir = 'logs'  
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
# to create log files
# tensorboard_callback helps to let developer come back and see how model performed where did it lag
# where we need to change parametrs n all
```

```python
hist = model.fit(train, epochs=20,validation_data=val,callbacks=[tensorboard_callback])
#callback transfer all of the info to the tensorboard
```

# Evaluation Of Model

### Evaluate

```python
from tensorflow.keras.metrics import Precision,Recall,BinaryAccuracy
```

```python
pre = Precision()
re = Recall()
acc = BinaryAccuracy()
```

```python
len(test)
```

```python
for batch in test.as_numpy_iterator():
    X,y = batch #x-images y-true value
    yhat = model.predict(X) #yhat = predicted value
    pre.update_state(y,yhat) #comparing og value and predicted value
    re.update_state(y,yhat)
    acc.update_state(y,yhat)
```

```python
print(f'Precision:{pre.result().numpy()}, Recall:{re.result().numpy()} , Accuracy:{acc.result().numpy()}')
```
